```python
tinggi = float(input("Masukkan tinggi badan (meter): "))
bmi = float(input("Masukkan BMI yang diharapkan: "))

berat = bmi * (tinggi ** 2)

print("Berat badan yang diperlukan adalah", berat, "kg")
```

```python
x = int(input("Masukkan nilai x (bilangan bulat): "))

f = 2*(x**3) + 2*x + 15/x

print("Nilai f(x) adalah:", f)
```

```python
gaji = float(input("Masukkan gaji per jam: "))
jam = float(input("Masukkan jam kerja per minggu: "))

kotor = gaji * jam * 5
bersih = kotor - (0.14 * kotor)

baju = 0.10 * bersih
alat = 0.01 * bersih

sisa = bersih - baju - alat
sedekah = 0.25 * sisa

yatim = 0.30 * sedekah
dhuafa = sedekah - yatim

print("Pendapatan sebelum pajak:", kotor)
print("Pendapatan setelah pajak:", bersih)
print("Untuk baju & aksesoris:", baju)
print("Untuk alat tulis:", alat)
print("Total sedekah:", sedekah)
print("Untuk anak yatim:", yatim)
print("Untuk kaum dhuafa:", dhuafa)
```
