```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y, 'b',x,z,'r')
plt.xlabel('Radians');
plt.ylabel('Value');
plt.title('Plotting Demonstration')
plt.legend(['sin', 'Cos'])
plt.grid()
```

```python
# Definisi variabel awal
jumlah_emas = 25
harga_beli = 650000
harga_jual = 685000

# Hitung keuntungan dalam rupiah
total_modal = jumlah_emas * harga_beli
total_nilai = jumlah_emas * harga_jual

keuntungan_rupiah = total_nilai - total_modal

# Hitung keuntungan dalam persentase
keuntungan_persen = (keuntungan_rupiah / total_modal) * 100

print(f"Keuntungan rupiah: {keuntungan_rupiah}")
print(f"Keuntungan persen: {keuntungan_persen}%")
```

```python
# Pembelian emas tambahan
emas_tambahan = 15
harga_beli_tambahan = 685000

# Total kepemilikan emas dan modal
total_emas_dimiliki = jumlah_emas + emas_tambahan
total_modal_keseluruhan = (jumlah_emas * harga_beli) + (emas_tambahan * harga_beli_tambahan)

# Harga emas terbaru
harga_emas_terkini = 715000
total_nilai_emas = total_emas_dimiliki * harga_emas_terkini

# Hitung keuntungan akhir
keuntungan_rupiah_total = total_nilai_emas - total_modal_keseluruhan
keuntungan_persen_total = (keuntungan_rupiah_total / total_modal_keseluruhan) * 100

print(f"Total modal: {total_modal_keseluruhan}")
print(f"Keuntungan rupiah: {keuntungan_rupiah_total}")
print(f"Keuntungan persen: {keuntungan_persen_total}%")
```
